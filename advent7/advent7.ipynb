```python
from operator import itemgetter
```

```python
def substitue_letter_by_val(hand):
    hand =[*hand]
    for k,v in [('T', 10), ('J', 11), ('Q', 12), ('K', 13), ('A', 14)]:
        hand = [v if h == k else h for h in hand]

    hand = [int(h) for h in hand]
    return hand
```

```python
with open('advent7.txt', 'r') as f:
    filedata = f.read()

hands = []
bids = []

for game in filedata.split('\n'):
    hand, bid = game.split(' ')
    hands.append(hand)
    bids.append(int(bid))
```

```python
counts = []
counts_bids = []
for hand, bid in zip(hands, bids):
    card_count = {'A': 0, 'K': 0, 'Q': 0, 'J': 0, 'T': 0, '9': 0, '8': 0, '7': 0, '6': 0, '5': 0, '4': 0, '3': 0, '2': 0, '1': 0}

    for key in card_count.keys():
        card_count[key] = hand.count(key)

    card_count = {k:val for k,val in card_count.items() if val != 0}
    card_count = sorted(card_count.values())[::-1]
    counts_bids.append((card_count, substitue_letter_by_val(hand), bid))

counts_bids = sorted(counts_bids,  key=(itemgetter(0,1)), reverse=True)
```

```python
counts_bids
```

```python
sum = 0
for i, (_, _, bid) in enumerate(counts_bids):
    sum += bid * (len(counts_bids) - i)

print(sum)
```

```python
def substitue_letter_by_val(hand):
    hand =[*hand]
    for k,v in [('T', 10), ('J', 0), ('Q', 12), ('K', 13), ('A', 14)]:
        hand = [v if h == k else h for h in hand]

    hand = [int(h) for h in hand]
    return hand

counts = []
counts_bids = []
for hand, bid in zip(hands, bids):
    card_count = {'A': 0, 'K': 0, 'Q': 0, 'J': 0, 'T': 0, '9': 0, '8': 0, '7': 0, '6': 0, '5': 0, '4': 0, '3': 0, '2': 0, '1': 0}

    for key in card_count.keys():
        card_count[key] = hand.count(key)

    jokers = card_count['J']
    
    if jokers != 0 and jokers != 5:
        del card_count['J']

    card_count = {k:val for k,val in card_count.items() if val != 0}
    try:
        if jokers != 5:
            card_count[max(card_count, key=card_count.get)] += jokers
    except:
        pass

    # card_count['J'] = jokers
    card_count = sorted(card_count.values())[::-1]
    counts_bids.append((card_count, substitue_letter_by_val(hand), hand, bid))

counts_bids = sorted(counts_bids,  key=(itemgetter(0,1)), reverse=True)
```

```python
sum = 0
for i, (_, _,_, bid) in enumerate(counts_bids):
    sum += bid * (len(counts_bids) - i)

print(sum)
```
